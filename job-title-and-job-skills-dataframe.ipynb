```python
job_skills = pd.read_csv("/kaggle/input/1-3m-linkedin-jobs-and-skills-2024/job_skills.csv")
linkedin_job_postings = pd.read_csv("/kaggle/input/1-3m-linkedin-jobs-and-skills-2024/linkedin_job_postings.csv")
```

```python
job_skills.columns
```

```python
linkedin_job_postings.columns
```

```python
# Assuming `job_skills` and `linkedin_job_postings` are your DataFrames
merged_df = linkedin_job_postings.merge(job_skills, on='job_link', how='inner')

# Select only the desired columns
job_and_skills_df = merged_df[['job_title', 'job_skills']]

# Display the resulting DataFrame
job_and_skills_df.head()
```

```python
# Convert the DataFrame to a CSV file
job_and_skills_df.to_csv("job_and_skills.csv", index=False)

print("DataFrame successfully saved to 'job_and_skills.csv'")
```
